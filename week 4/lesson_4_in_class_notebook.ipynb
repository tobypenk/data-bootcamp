```python
df["investors"] = df["investors"].str.split(",")
df
```

```python
new_df = []

for i in range(len(df)):

  for j in df.iloc[i]["investors"]:
    new_row = [
        df.iloc[i]["col1"],
        df.iloc[i]["col2"],
        df.iloc[i]["col3"],
        j
    ]
    new_df.append(new_row)
```

```python
new_df = pd.DataFrame(new_df,columns=["col1","col2","col3","investor"])
new_df
```
