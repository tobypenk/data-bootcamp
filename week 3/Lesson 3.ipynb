## Classes

Let's create a little game. In the game, the user will design an animal. They will design the animal by choosing values for these qualities, or `attribute`s:

- A name (str)
- A greeting like "woof" (str)
- A number of legs (int)
- A number of arms (int)
- A stash of food, in pounds (float)
- A daily requirement for food, in pounds (float)


This isn't too hard to implement with our current knowledge of coding. We can just declare a few variables:

    name = "Mushroom"
    greeting = "Meow"
    legs = 4
    arms = 0
    food_stash_lbs = 10
    daily_food_need_lbs = 0.25
    
Now we can make the animal do stuff. How about the animal eats some food? In our game, we could think of this as the equivalent of ticking forward one day in the simulated animal's life. We might want to do this repeatedly, so we make a function we can call repeatedly:

    def eat_food(amount):
        global food_stash_lbs
        food_stash_lbs -= amount

A new keyword has entered the stage: `global`. This allows the function to talk to the `food_stash_lbs` variable and decrement it even though it is not naturally in the function's scope. Remember though that having functions interact with global-scope variables is not considered a great coding practice - reaching "up" a scope level is unexpected behavior that can cause your scripts to behave weirdly and be hard to debug. But this works for our purposes, for now.

Now we have a new user who wants to play the game, and they want to create a new animal named `"Anubis"` who will say `"Woof"`. How are we going to implement this without removing `"Mushroom"` from the system? We could rename all the variables:
    
    mushroom_name = "Mushroom"
    mushroom_greeting = "Meow"
    mushroom_legs = 4
    mushroom_food_stash_lbs = 10
    mushroom_daily_food_need_lbs = 0.25
    
    anubis_name = "Anubis"
    anubis_greeting = "Woof"
    anubis_legs = 4
    anubis_food_stash_lbs = 40
    anubis_daily_food_need_lbs = 1.25
    
We'll also need to define custom functions for each one to decrement their food. This works too, but it's going to get really annoying and messy, really fast. It's already pretty annoying!

Fortunately there is a different way: `class`es. The system of coding we've been working in so far is called _functional_ programming, and it is focused on the relationship between _inputs_ and _outputs_ of functions and data, and the _transformations_ that turn one into the other. The alternative paradigm is called _object-oriented_ programming, and in it, code behavior is mediated by _classes_, which take on semantic meaning similar to how a human might think of objects in the real world. This will make more sense in a second.

Functional and object-oriented programming are both useful, and one is not better than the other - though one might be a more natural fit for certain tasks. Often the choice between the two will be simply one of preference or your comfort level.

To create the same class as we did above, we'd use this notation:

    class Animal:
    
        def __init__(self,name,greeting,legs,food_stash_lbs,daily_food_need_lbs):
            self.name = name
            self.greeting = greeting
            self.legs = legs
            self.food_stash_lbs = food_stash_lbs
            self.daily_food_need_lbs = daily_food_need_lbs

        def eat_food(self):
            self.food_stash_lbs -= self.daily_food_need_lbs
            print("munch munch")
            
    mushroom = Animal("Mushroom","Meow",4,10,0.25)
        
Note a few things here:
- The `class` keyword is to classes what `def` is to functions
- The `__init__` function (with _two_ underscores on each side) is necessary for _initializing_ the class
    - _inside_ the `__init__` function, we take all the inputs and assign them to `self.[whatever]` - we could also do preprocessing here, like `self.legs_per_side = int(legs / 2)`
- Every function, including `__init__`, receives `self` as a parameter
    - However, notice that in creating the actual _instantiation_ of mushroom, below the class declaration, that I _do not_ pass anything for the `self` variable. It is ignored in instantiation.
- The first letter of the class name is _uppercased_ - this is not strictly required but is very common practice

How do we make `mushroom` do stuff? Just call her methods like this:

    mushroom.eat_food()
    
Notice again, I don't have to pass anything for the `self` parameter. How do we find out mushroom's current level of food?

    print(mushroom.food_stash_lbs)
    
Maybe we want to add a `method` instead - we can add this function to the inside of the class declaration:

    def food_status(self):
        days_left = self.food_stash_lbs / self.daily_food_need_lbs
        s1 = "I have " + str(self.food_stash_lbs) + " lbs of food left."
        s2 = "My food will last me another " + str(days_left) + " days."
        return " ".join([s1,s2])
        
Notice that within the class, the class's own variables are prefaced with `self.` so Python knows to look for them _inside_ the class and to use only the ones that are relevant to the specific instance you've created.

Class methods can also take new parameters from the outside world. Add this to the inside of the class declaration:

    def greet(self,who):
        return self.greeting + " " + who + "!"
        
We call this with:

    mushroom.greet("Toby")
    
Notice again, we ignore the `self` parameter when _calling_ the function, even though it is required in the _declaration_.

## Handling errors

Sometimes you will write code that you think might return an error - say, if someone tries to pass a `str` where an `int` should be. But when Python throws an error, it doesn't just keep going - it terminates execution of all the rest of your code. Take this example:

    def divide_by_two(i):
        return i / 2
    
    divide_by_two("3")
    print("program done")

The function errors out during the call to `divide_by_two` and the final line is never printed. How can errors be handled more gracefully?

### try-except

Using Python's `try-except` pattern, you can test for errors and handle them gracefully:

    def graceful_divide_by_two(i):
        try:
            return i / 2
        except:
            return "there was a problem with the operation - check your data type"

    print(graceful_divide_by_two("3"))
    print("program done")
    
Now the program will print the much more helpful error message, _and_ finish running the rest of your script. `try-except` is critical for circumstances where errors might happen, but the rest of your script needs to finish running. Other times, you will want the script to fail so erroneous output doesn't get created at all. The decision is based on context and the specific needs of your program.

## List comprehensions

Recall the concept of _syntactic sugar_ - design pattersn Python makes available to make some tasks easier to write with less code. For example, these are equivalent:

    # normal way
    if 50425 % 3 == 0:
        x = "yes"
    else:
        x = "no"
        
    # sugary way (ternary operator)
    x = "yes" if 50425 % 3 == 0 else "no"
    
Syntactic sugar gives coders tools that make the experience of coding feel smoother and helps us get in, and maintain, a flow state while coding. Typically it also reduces lines or characters of code. It should be used thoughtfully because often the sugary way of writing something is harder to read.

`list`s have syntactic sugar also. Specifically, we can `map` and `filter` lists without calling the `map` and `filter` functions by using alternative notation. These are equivalent:

    # long way of filtering
    ints = list(range(200))
    multiples_of_7 = []
    for i in ints:
        if i % 7 == 0:
            multiples_of_7.append(i)

    # "pythonic" way of filtering
    ints = list(range(200))
    multiples_of_7 = list(filter(lambda x: x % 7 == 0, ints))
    
    # sugary way (list comprehension)
    ints = list(range(200))
    multiples_of_7 = [x for x in ints if x % 7 == 0]
    
These three methods produce equivalent outputs, but are in descending order of number of characters written.

List comprehensions can also be used to _map_ lists:

    # long way of mapping
    ints = list(range(20))
    for i in range(len(ints)):
        ints[i] *= 7

    # "pythonic" way of mapping
    ints = list(range(20))
    multiplied_by_7 = map(filter(lambda x: x * 7, ints))
    
    # sugary way (list comprehension)
    ints = list(range(20))
    multiplied_by_7 = [x * 7 for x in ints]
    
The general rule is this: [_expression_ for _iterator_ in _iterable_ if _condition_ ]. You just directly wrap the expression in square brackets. Your _iterator_ can be called whatever you want, subject to the usual variable-naming rules. Your iterable needs to be some kind of collection of things like a list.

You can, of course, also combine mapping and filtering:

    ints = list(range(200))
    squared_multiples_of_7 = [x ** 2 for x in ints if x % 7 == 0]
    
It's worth repeating: be wary of using list comprehensions for things that are more complex. They can get hard to read:

    # get squared factorials of multiples of 7 or 9 or anything under 10
    import functools
    ints = list(range(1,200))
    new_list = [functools.reduce(lambda i,j: i*j,list(range(1,x+1)))**2 for x in ints if x % 7 == 0 or x % 9 == 0 or x < 10]
    
Everyone would probably be better off if we just wrote this with more lines of code.

## Dealing with files

Frequently you will need to open, manipulate, and write files with your code. If the files are csv's, you will usually want to use the built-in functions in `pandas` to handle them; however, there are other times when you will want to load in raw data. Python has a simple way to do this:

    # one way considered the most "pythonic"
    with open("file_1.txt") as file:
        file_data = file.read()
        
    # another exactly equivalent way that might make more sense at first
    file = open("file_1.txt")
    file_data = file.read()
    
In these examples, the `file` variable becomes a _file object_, which has methods such as `read`, `write`, and `close`.

`open()` will first look to see if the file you referenced exists. If the file doesn't exist, it will be created; if it does already exist, it will be read.

The _file object_ does not directly contain the data in the file; to get the data, you must call `read()` or `readline()`. They work similarly, except that `read()` takes in the entire file, while `readline()` reads one line at a time:

    file = open("made_up_file.txt")
    file.read()
        "line 1"
        "line 2"
        "line 3"
        
    file = open("made_up_file.txt")
    file.readline()
        "line 1"
    file.readline()
        "line 2"
    file.readline()
        "line 3"
        
You can also use a `for` loop to read all the lines in a file:

    file = open("made_up_file.txt")
    for line in file:
        print(line)
        
Finally, make sure to `close` your file objects:

    file.close()
    
If you leave the file connection open in your workspace, you can unintentionally modify it later, with painful and annoying results.


### Writing files

If you plan to `write` data to a file, you need to choose the appropriate flag when you `open` the file:

    # open the file for reading only (this is the default)
    file = open("test_file.txt","r")
    
    # open the file for writing, overwriting the file when you write
    file = open("test_file.txt","w")
    
    # open the file for writing, appending to the existing data
    file = open("test_file.txt","a")

There are other flags, such as `"x"`, which will cause `open()` to fail if the file already exists, but the above are the most commonly used ones.

`write`ing to the file is as simple as calling the `write()` function on the `file` object:

    file.write("some new data")
    
As always, make sure to `close()` the file object when you're done with it.


## Timing and efficiency

Sometimes you may wonder how long some code you've written takes or will take to execute. For this you can use the `time` library, which has a function (also called `time`) that logs the current system time with extreme precision. By logging the current time, running some code, logging the time again, and finding the difference between the two, you can determine how long the code took.

    import time
    
    time_before = time.time()
    for i in range(1000):
        for j in range(1000):
            x = i * j
    time_after = time.time()
    time_elapsed = time_after - time_before

This concept is useful when you're either:
1. Curious how long code you've written is taking to execute
2. Want to iterate _part_ of a loop, say 1% of it, to get a sense of how long the full loop will take

### Complexity

Code efficiency is a big topic, and we'll only spend a little time on it here. For now, just remember that the concept of _"Big O"_ notation refers to code complexity, which is about how many computations your code will have to do in a worst-case scenario.

Say you have a list of 100 integers. Here is how "Big O" notation views the complexity of each of these operations:

    my_list = range(100)
    
    def constant_time(l):
        """
            this function runs in "constant time," also referred to as "O(1)"
            because it will take the same amount of time no matter how long
            the list is
        """
        return len(l)
        
    def linear_time(l):
        """
            this function is of "O(n)" complexity: its timing is proportional to
            n, which refers to the number of items in the list
        """
        for i in l:
            pass
            
    def square_time(l):
        """
            this function is of "O(n^2)" complexity: its timing is proportional to
            n squared, because for each element of l, it must loop the entire list.
            similar logic extends to O(n^3), etc.
        """
        for i in l:
            for j in l:
                pass
                
    def still_linear_time(l):
        """
            though this function involves more stuff than the `linear_time` function,
            it is still considered of O(n) complexity. the number of calculations here
            is actually 3 * n, but constants (like the 3 in this case) are omitted, as
            over large datasets, the size of the data enormously outweighs the number
            of calculations in terms of time drag.
        """
        for i in l:
            x = i ** 2 * 3 + 4
            
    def still_linear_time_again(l):
        """
            this function performs n / 2 iterations, but by the same logic above, is 
            considered of O(n) complexity.
        """
        for i in range(len(l) // 2):
            pass
            
    def log_time(l,search_val):
    """
        this function performs a really cool operation called binary search. in the
        worst-case scenario, binary search only needs to perform log2(n) calculations,
        so it is considered of O(log n) complexity. this is powerful because a list 
        with twice as many elements as another will take only ONE more operation to 
        search. binary search can be used to search a sorted list of a quadrillion 
        elements for a specific value in just 40 operations, the blink of an eye.
        
        search is a huge topic which we'll mostly stay away from in this course, but 
        it is the motivation for a lot of complexity analysis.
    """
    
    min_idx = 0
    max_idx = len(l)
    idx = len(l) // 2
    
    while l[idx] != search_val:
        if l[idx] > search_val:
            max_idx = idx
            idx = (idx + min_idx) // 2
        else:
            min_idx = idx
            idx = (idx + max_idx) // 2
    return idx
         
There are lots more algorithms that lead to interesting complexities: square root complexity, factorial complexity (very bad), and others. The bottom line: if your code is slow, think about how many operations it's doing for a hint at where you might be able to optimize it.

## Pandas

Pandas is a Python library used to manipulate and analyze data. It is the primary tool you will likely use for both your midterm and your final projects.

The core data types in pandas are:
- Series: one-dimensional datasets, much like lists. Series can contain any type of data, including mixed data in a single Series.
- DataFrames: two-dimensional datasets, much like spreadsheets. Each column in a DataFrame is a Series, so DataFrames can be thought of as wrappers for many Series.

One way to create a DataFrame is to use a list of lists, and then pass that list of lists to the DataFrame constructor:
    
    import pandas as pd
    x = [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ]

This will give back a DataFrame that looks exactly like the grid of numbers: the first column is 1, 4, 7, etc. If we needed the first column to be 1, 2, and 3, we'd `transpose` the DataFrame using `df.transpose()`.

Currently, our columns and rows don't have names; they just have zero-indexed numbers. The first (or zeroth) column can be accessed by `df[0]`, and the first _value_ in the first row can be accessed by `df[0][0]`. If you want to name the columns and rows to make them more readable, you can use the `columns` and `index` parameters during creation:
    
    import pandas as pd

        [1,2,3],
        [4,5,6],
        [7,8,9]
    ]
    df = pd.DataFrame(x,columns=["a","b","c"],index=["x","y","z"])

Now, the columns are named a, b, and c, while the rows are named x, y, and z. So now accessing the first value in the first column looks like `df["a"]["x"]`.

Generally you will want to use `loc` or `iloc` instead of brackets, because these functions are more _performant_, meaning faster / more efficient, for reasons under the hood of Python and your machine.

### Slicing

Often you will want more than one piece of data. This is easy using the colon symbol:
    
    x = [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ]
    df = pd.DataFrame(x)
    df.iloc[0:2]
    
This gives the first and second columns as the indexing stops one _before_ the higher number. With a comma, you can select slices of both columns _and_ rows:

    df.iloc[0:2,0:2]
    
This gives the first two rows of the first two columns.

If you want to use `loc` instead of `iloc`, and your columns are named, you can provide a `list` of values:

    x = [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ]
    df = pd.DataFrame(x,columns=["a","b","c"],index=["x","y","z"])
    df.loc[["x","y"],["a","b"]]
    
Note that brackets select columns first, then rows: `df[column][row]`, whereas `loc` and `iloc` select rows, then columns: `df.iloc[row,column]`. `loc` and `iloc` use matrix conventions.
    


### Filtering

What if you need to filter a dataframe based on some more interesting condition than just a continuous selection of rows or columns? You can provide a _boolean mask_ to accomplish this:

    x = [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ]
    df = pd.DataFrame(x)
    mask = df[0] % 2 == 0
    df.loc[mask]
    
What's going on here? You create a mask testing for evenness (aka divisibility by 2) of the first element of each row - this mask takes on 3 values, one for each row. The values, in order, are False, True, False. pandas holds that mask up to the dataframe and retains only the rows that have a `True` value next to them - in this case, just row index 1. If you changed the criterion:

    mask = df[1] % 2 == 0
    df.loc[mask]
    
Now you're selecting for evenness in the _second_ column `df[1]` - so the first and last row are retained because they have the values 2 and 8 in their second column, respectively. Masks can also be applied to columns:

    mask = df.loc[0] == 3
    df.loc[:,mask]
