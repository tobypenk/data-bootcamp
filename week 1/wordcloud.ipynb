```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("beginning survey.csv")
```

```python
text = " ".join(df["What 3 things do you hope to get from this class?"]).lower()
```

```python
word_cloud = WordCloud(
    collocations = False, 
    background_color = 'white').generate(text)
```

```python
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
