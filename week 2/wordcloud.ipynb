```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("20230202 beginning of term survey.csv")
```

```python
text = " ".join(df["What 3 things do you hope to get from this class?"]).lower()
text = text.replace("python","")
text = text.replace("data","")
text = text.replace("learn","")
```

```python
word_cloud = WordCloud(
    collocations = False, 
    background_color = 'white'
).generate(text)
```

```python
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```
